```python
from agentic_doc.parse import parse
```

```python
result = parse("../data/raw/HSC26-Bangla1st-Paper.pdf", include_marginalia=False,
               include_metadata_in_markdown=False,)
```

```python
splits = [
    chunk for chunk in result[0].chunks if chunk.chunk_type in ("text", "table")]
```

```python
from langchain_core.documents import Document
documents = []
for split in splits:
    if split.chunk_type == "text" or split.chunk_type == "table":
        documents.append(Document(
            page_content=split.text,
            metadata={
                'id': split.chunk_id,
            }
        ))
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="intfloat/multilingual-e5-large-instruct")
```

```python
from langchain_community.vectorstores import Chroma
```

```python
vectorstore = Chroma.from_documents(documents=documents,
                                    embedding=embeddings)

retriever = vectorstore.as_retriever()
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="moonshotai/kimi-k2-instruct",
)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


template = """Answer the question based only on the following context:
{context}

Question: {query}
"""
prompt = ChatPromptTemplate.from_template(template)

qa_chain = prompt | llm | StrOutputParser()
```

```python
def format_docs(relevant_docs):
    return "\n".join(doc.page_content for doc in relevant_docs)


query = "ржЕржирзБржкржорзЗрж░ ржнрж╛рж╖рж╛рзЯ рж╕рзБржкрзБрж░рзБрж╖ ржХрж╛ржХрзЗ ржмрж▓рж╛ рж╣рзЯрзЗржЫрзЗ?"

relevant_docs = retriever.invoke(query)
qa_chain.invoke({"context": format_docs(relevant_docs), "query": query})
```

```python
docs_with_score = vectorstore.similarity_search_with_score(query)
```
