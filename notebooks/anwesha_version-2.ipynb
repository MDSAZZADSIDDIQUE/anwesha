```python
from agentic_doc.parse import parse
```

```python
result = parse("../data/raw/HSC26-Bangla1st-Paper.pdf")
```

```python
result[0].markdown
```

```python
splits = [
    chunk for chunk in result[0].chunks if chunk.chunk_type in ("text", "table")]
```

```python
splits[0]
```

```python
from langchain_core.documents import Document
documents = []
for split in splits:
    print()
    if split.chunk_type == "text" or split.chunk_type == "table":
        documents.append(Document(
            page_content=split.text,
            metadata={
                'id': split.chunk_id,
            }
        ))
```

```python
from langchain_community.vectorstores import Chroma
```

```python
import getpass
import os

if not os.getenv("GOOGLE_API_KEY"):
    os.environ["GOOGLE_API_KEY"] = getpass.getpass(
        "Enter your Google API key: ")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="intfloat/multilingual-e5-large-instruct")
```

```python
documents
```

```python
vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    persist_directory="../database/vectorstore.db"
)

retriever = vectorstore.as_retriever()
```
