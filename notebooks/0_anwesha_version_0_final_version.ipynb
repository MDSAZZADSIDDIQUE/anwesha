```python
import re
import unicodedata

COMPILED_BOILERPLATE_PATTERNS = [
    re.compile(p, flags=re.MULTILINE | re.IGNORECASE) for p in [
        r"HSC\s?'?26\s+ржЕржирж▓рж╛ржЗржи\s+(?:ржмрзНржпрж╛ржЪ|ржЧрж╛ржЗржб|рж╕рзНржХрзБрж▓|ржХрзНржпрж╛ржЯ)",
        r"ржмрж╛ржВрж▓рж╛\s*[тАв-]\s*ржЗржВрж░рзЗржЬрж┐\s*[тАв-]\s*(?:ржЖржЗрж╕рж┐ржЯрж┐|ржЧржгрж┐ржд|рждржерзНржп\s+ржУ\s+ржпрзЛржЧрж╛ржпрзЛржЧ(?: ржкрзНрж░ржпрзБржХрзНрждрж┐)?)",
        r'10\s*MINUTE\s*SCHOOL',
        r'ржЕржирж▓рж╛ржЗржи ржмрзНржпрж╛ржЪ рж╕ржорзНржкрж░рзНржХрж┐ржд ржпрзЗржХрзЛржирзЛ ржЬрж┐ржЬрзНржЮрж╛рж╕рж╛ржпрж╝,',
        r'ржХрж▓ ржХрж░рзЛ\s*ЁЯУЮ?\s*16910',
        r'RESERVED ERN BENGAL HALWA',
        r'---',
        r'^\d+$',
        r'SL\s*Ans',
        r'</?table\b[^>]*>', r'</?tr\b[^>]*>', r'</?td\b[^>]*>',
        r'</?thead\b[^>]*>', r'</?tbody\b[^>]*>',
        r'\[Image \d+\]',
        r'^\s*[,."]*\s*$',
    ]
]


def _remove_boilerplate_and_artifacts(text: str) -> str:
    for pattern in COMPILED_BOILERPLATE_PATTERNS:
        text = pattern.sub('', text)
    return text


def _correct_ocr_and_spelling(text: str) -> str:
    correction_map = {
        'ржЕржирж▓рж╛ржЗржи ржХрзНржпрж╛ржЯ': 'ржЕржирж▓рж╛ржЗржи ржмрзНржпрж╛ржЪ', 'ржХржирзНржпрж╛ржЧрзА': 'ржХрж▓рзНржпрж╛ржгрзА', 'рж╢рждрзНрж░рзБржирж╛рж╢': 'рж╢ржорзНржнрзБржирж╛рже',
        'рж╢рзБржирзБрж░рж╛ржЦржмрж╛рж░рзБ': 'рж╢ржорзНржнрзБржирж╛ржержмрж╛ржмрзБ', 'рж╢рзБржнржирж╛рже': 'рж╢ржорзНржнрзБржирж╛рже', 'ржкржгрзНржбрж┐рждржорж╢рж╛рзЯ': 'ржкржгрзНржбрж┐рждржорж╢рж╛ржЗ',
        'ржЧрзБржирзЗрж░': 'ржЧрзБржгрзЗрж░', 'ржЧрзБржи': 'ржЧрзБржг', 'рж░рзБржк': 'рж░рзВржк', 'ржмрж┐ржжрзНрж░рзЛржк': 'ржмрж┐ржжрзНрж░рзВржк', 'ржорзБрж╣рзБрж░рзНржд': 'ржорзБрж╣рзВрж░рзНржд',
        'ржжрзБржХрзНрж╖рзБрж░': 'ржжрзБржХрзНрж╖', 'ржпрзМрждрзВржХ': 'ржпрзМрждрзБржХ', 'ржмрзНржпрж╛ржХрзНрждрж┐рждрзНржмрж╣рзАржи': 'ржмрзНржпржХрзНрждрж┐рждрзНржмрж╣рзАржи',
        'ржХрж╛рж╣рж┐ржирзА': 'ржХрж╛рж╣рж┐ржирж┐', 'рж╢рзНрж░рзЗржгрж┐': 'рж╢рзНрж░рзЗржгрзА', 'ржмрж╛рзЬрзА': 'ржмрж╛рзЬрж┐', 'ржЧрж╛рзЬрзА': 'ржЧрж╛рзЬрж┐',
        'рж╢рж╛рзЬрзА': 'рж╢рж╛рзЬрж┐', 'ржКржкрж░': 'ржЙржкрж░', 'ржорж╛рж╕рж┐-ржкрж┐рж╕рж┐': 'ржорж╛рж╕рж┐ржкрж┐рж╕рж┐',
        'ржХ рзН рж╖': 'ржХрзНрж╖', 'ржЬ рзН ржЮ': 'ржЬрзНржЮ', 'ржЮ рзН ржЪ': 'ржЮрзНржЪ', 'ржЩ рзН ржХ': 'ржЩрзНржХ', 'ржЩ рзН ржЧ': 'ржЩрзНржЧ',
        'ржи рзН ржб': 'ржгрзНржб', 'ржи рзН ржд': 'ржирзНржд', 'ржд рзН ржд': 'рждрзНржд', 'ржд рзН рже': 'рждрзНрже', 'рж╢ рзН ржЪ': 'рж╢рзНржЪ',
        'рж╣ рзН ржо': 'рж╣рзНржо', 'рж╣ рзН ржи': 'рж╣рзНржи',
    }
    for wrong, right in correction_map.items():
        text = text.replace(wrong, right)
    return text


def _normalize_punctuation(text: str) -> str:
    text = re.sub(r'(?<!\d)\.(?!\d)\s*$', r'ред', text, flags=re.MULTILINE)
    text = re.sub(r'тЬУ\s*', r'* ', text)
    return text


def _rejoin_broken_words(text: str) -> str:
    text = re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)
    text = re.sub(r'([a-z_A-Z_ржЕ-рж╣])\n([a-z_A-Z_ржЕ-рж╣])', r'\1\2', text)
    return text


def _remove_zero_width_chars(text: str) -> str:
    return re.sub(r'[\u200b-\u200d\ufeff]', '', text)


def _normalize_whitespace(text: str) -> str:
    text = re.sub(r'[ \t]+', ' ', text)
    text = re.sub(r'\n{3,}', '\n\n', text)
    text = '\n'.join([line.strip() for line in text.split('\n')])
    return text.strip()


def clean_raw_markdown(raw_text: str) -> str:
    text = _rejoin_broken_words(raw_text)
    text = _remove_boilerplate_and_artifacts(text)
    text = unicodedata.normalize('NFC', text)
    text = _remove_zero_width_chars(text)
    text = _correct_ocr_and_spelling(text)
    text = _normalize_punctuation(text)
    text = _normalize_whitespace(text)
    return text
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from langchain.prompts import ChatPromptTemplate

template = """You are an AI language model assistant. Your task is to generate five 
different versions of the given user question to retrieve relevant documents from a vector 
database. By generating multiple perspectives on the user question, your goal is to help
the user overcome some of the limitations of the distance-based similarity search. 
Provide these alternative questions separated by newlines. Original question: {question}"""
prompt_perspectives = ChatPromptTemplate.from_template(template)


generate_queries = (
    prompt_perspectives
    | ChatGroq(model="moonshotai/kimi-k2-instruct",)
    | StrOutputParser()
    | (lambda x: x.split("\n"))
)
```

```python
from langchain.load import dumps, loads


def get_unique_union(documents: list[list]):
    flattened_docs = [dumps(doc) for sublist in documents for doc in sublist]
    unique_docs = list(set(flattened_docs))
    return [loads(doc) for doc in unique_docs]


question = "ржЕржирзБржкржо рждрж╛рж░ ржорж╛ржорж╛рж░ ржЪрзЗржпрж╝рзЗ ржХржд ржмржЫрж░рзЗрж░ ржЫрзЛржЯ ржЫрж┐рж▓?"
retrieval_chain = generate_queries | retriever.map() | get_unique_union
docs = retrieval_chain.invoke({"question": question})
len(docs)
```

```python
from langchain.load import dumps, loads


def reciprocal_rank_fusion(results: list[list], k=60):

    fused_scores = {}

    for docs in results:
        for rank, doc in enumerate(docs):
            doc_str = dumps(doc)
            if doc_str not in fused_scores:
                fused_scores[doc_str] = 0
            previous_score = fused_scores[doc_str]
            fused_scores[doc_str] += 1 / (rank + k)

    reranked_results = [
        (loads(doc), score)
        for doc, score in sorted(fused_scores.items(), key=lambda x: x[1], reverse=True)
    ]

    return reranked_results


question = "ржЕржирзБржкржо рждрж╛рж░ ржорж╛ржорж╛рж░ ржЪрзЗржпрж╝рзЗ ржХржд ржмржЫрж░рзЗрж░ ржЫрзЛржЯ ржЫрж┐рж▓?"
retrieval_chain_with_reranking = generate_queries | retriever.map() | reciprocal_rank_fusion
docs = retrieval_chain_with_reranking.invoke({"question": question})
len(docs)
```
